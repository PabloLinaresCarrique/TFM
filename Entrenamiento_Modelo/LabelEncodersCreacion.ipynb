## Premisa
Se realiza un codigo para generar los labelencoders que utilizaremos en el futuro

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle
import gdown
from google.colab import files
```

```python
# Descargar archivos desde Google Drive
hi_small_id = '1fDw0plQy898cw5aPZ8qBfP61djO6ssDH'
low_small_id = '1LIGRW_huhOGhS1Xl1n-v39R7KweRYMny'
low_medium_id = '1u3N2VHM2nyso33F5hgE4zxDuMv09KQ-w'

gdown.download(f'https://drive.google.com/uc?export=download&id={hi_small_id}', 'HI-Small_Trans.csv', quiet=False)
gdown.download(f'https://drive.google.com/uc?export=download&id={low_small_id}', 'LI-Small_Trans.csv', quiet=False)
gdown.download(f'https://drive.google.com/uc?export=download&id={low_medium_id}', 'HI-Medium_Trans.csv', quiet=False)

# Cargar los datasets
hi_small = pd.read_csv('HI-Small_Trans.csv')
low_small = pd.read_csv('LI-Small_Trans.csv')
low_medium = pd.read_csv('HI-Medium_Trans.csv')
```

```python
hi_small.rename(columns={'Account': 'Account2', 'Account.1': 'Account4'}, inplace=True)
low_small.rename(columns={'Account': 'Account2', 'Account.1': 'Account4'}, inplace=True)
low_medium.rename(columns={'Account': 'Account2', 'Account.1': 'Account4'}, inplace=True)
```

```python
df_unido =  pd.concat([hi_small, low_small, low_medium], ignore_index=True)
```

```python
df_unido['Timestamp'] = pd.to_datetime(df_unido['Timestamp'])

df_unido['Year'] = df_unido['Timestamp'].dt.year
df_unido['Month'] = df_unido['Timestamp'].dt.month
df_unido['Day'] = df_unido['Timestamp'].dt.day
df_unido['Hour'] = df_unido['Timestamp'].dt.hour
df_unido['Minute'] = df_unido['Timestamp'].dt.minute

df_unido.drop(columns=['Timestamp'], inplace=True)
```

```python
categorical_columns = ['From Bank', 'To Bank', 'Receiving Currency', 'Payment Currency', 'Payment Format', 'Account2', 'Account4']
```

```python
label_encoders = {}
categorical_columns = ['From Bank', 'To Bank', 'Receiving Currency', 'Payment Currency', 'Payment Format', 'Account2', 'Account4']

# Aplicar LabelEncoder a las columnas categóricas
for col in categorical_columns:
    le = LabelEncoder()
    df_unido[col] = le.fit_transform(df_unido[col])
    label_encoders[col] = le

# Guardar el diccionario de label encoders localmente en Colab
output_file_path = 'label_encoders1.pkl'
with open(output_file_path, 'wb') as f:
    pickle.dump(label_encoders, f)

# Descargar el archivo directamente a tu equipo local
files.download(output_file_path)
```
